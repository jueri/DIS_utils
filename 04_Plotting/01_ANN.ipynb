04.01.01. *Accuracy* and *Loss* results from training by a *test_plan*
---

To learn about different hyperparameters of an ANN I trained a modle multiple times and changed one parameter according to a predefined list, the *test_plan*.

For example different batch sizes, defined as:
```
test_plan = [2,4,8,16,32,64,128,256,512,1024,2048]
```

````python
import matplotlib.pyplot as plt

def makeChart(title: str, lable: str, test_plan: list): 
  """
  This function provides a standardized Plot for comparing the latest results.
  The Plot is separated into two subplots, one for accuracy and one for the 
  loss score.

  Parameters
  ----------
  title : str
    The global title of the plot
  lable : str:
    The lable for the x axis
  test_plan : list
    The list of values for the x axis
  """
  plt.figure(figsize=(12,12))

  plt.subplot(221)
  plt.plot(test_plan, results_acc)
  plt.xlabel(lable)
  plt.ylabel('accuracy')
  plt.title('Accuracy')

  plt.subplot(222)
  plt.plot(test_plan, results_loss)
  plt.xlabel(lable)
  plt.ylabel('loss')
  plt.title('Loss')
  plt.suptitle(title,fontsize=20)

  plt.show()
````
